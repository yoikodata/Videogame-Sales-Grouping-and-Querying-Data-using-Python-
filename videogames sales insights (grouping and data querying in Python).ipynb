```python
#6
Electronic_Activision = videogames[(videogames['Publisher']=='Electronic Arts')|(videogames['Publisher']=='Activision')]
```

```python
Electronic_Activision[Electronic_Activision['Global_Sales'] > 5]
#All data for publisher EA and Activision with sales over $5M
```

```python
#7
condition1 = videogames['Publisher']=='Electronic Arts'
condition2 = videogames['Publisher']=='Activision'
condition3 = videogames['Publisher']=='Nintendo'
condition4 = videogames['Publisher']=='Ubisoft'
```

```python
df = videogames[condition1 | condition2 | condition3 |condition4] 
```

```python
df_over_5 = df[df['Global_Sales'] > 5]
```

```python
count = df_over_5.groupby('Publisher')['Name'].count()
```

```python
sum_sales = df_over_5.groupby('Publisher')['Global_Sales'].sum()
```

```python
pd.concat([count, sum_sales], axis=1)
#number of games and total global sales by Publisher
```

```python
#8
sales_by_publisher = videogames.groupby('Publisher')['Global_Sales'].sum()
```

```python
pct_sales_by_publisher = round(sales_by_publisher.div(sum(sales_by_publisher)), 2).sort_values(ascending=False)*100
```

```python
print(pct_sales_by_publisher.iloc[0:16])
#Top 15 Publishers based on percentage of global sales
```

```python
#9
sales_by_platform = videogames.groupby('Platform')['Global_Sales'].sum()
```

```python
pct_sales_by_platform = round(sales_by_platform.div(sum(sales_by_platform)), 2).sort_values(ascending=False)*100
```

```python
print(pct_sales_by_platform.iloc[0:16])
#Top 15 gaming systems based on percentage of global sales
```
